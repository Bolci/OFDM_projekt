```python
import string
import re
import random
import numpy as np
import torch
import torch.nn as nn
from torch import optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from dataclasses import dataclass
import itertools
import matplotlib.pyplot as plt
from copy import copy
from scipy.stats import entropy, rayleigh


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

## Dataset generation

```python
@dataclass
class QPSK_symbols:
    symbols = np.asarray([0, 1 + 1j, 1 - 1j, -1 + 1j, -1 - 1j])
    amplitude = np.sqrt(2)


@dataclass
class QAM_symbols:
    symbols = np.asarray([3 + 3j,
                          3 + 1j,
                          3 - 3j,
                          3 - 1j,
                          1 + 3j,
                          1 + 1j,
                          1 - 3j,
                          1 - 1j,
                          -3 + 3j,
                          -3 + 1j,
                          -3 - 3j,
                          -3 - 1j,
                          -1 + 3j,
                          -1 + 1j,
                          -1 - 3j,
                          -1 - 1j])
    amplitude = np.sqrt(18)


class OFDMDataset(Dataset):
  def __init__(self, n_of_data: int, nfft: int = 32, modulation_type:str ='qpsk', n_zero: int = 8, to_bins = False):
    super().__init__()

    self.nfft = nfft
    self.n_of_data = n_of_data
    self.modulation_type = modulation_type
    self.symbols = {'qpsk': QPSK_symbols, 'QAM': QAM_symbols}

    self.max_aplitude = self.symbols[modulation_type].amplitude
    self.symbols = self.symbols[modulation_type].symbols

    index_symbols = np.arange(1, len(self.symbols))
    self.generated_sequences_ids = np.random.choice(index_symbols, size=(n_of_data, nfft))

    self.complex_symbols = self.symbols[self.generated_sequences_ids]

    ifft_symbols = np.fft.ifft(np.asarray(self.complex_symbols))

    self.abs_values = np.abs(ifft_symbols)

    self.generated_sequences_ids = torch.tensor(self.generated_sequences_ids)

    self.abs_values = torch.tensor(self.abs_values, dtype=torch.float32)

    self.generated_sequences_ids_one_hot = F.one_hot(self.generated_sequences_ids, num_classes=len(self.symbols)).type(torch.float32)
    self.generated_sequences_ids = self.generated_sequences_ids.type(torch.float32)

    if to_bins:
      self.abs_values, _, _ = self.convert_to_vocab(self.abs_values)
  
  def convert_to_vocab(self, data, bins: int = 200) -> tuple:
    counts, bins, patches = plt.hist(data, bins=bins)
    new_vec = copy(data)

    for ids in range(len(bins)-1):
      i, j = np.where((data>=bins[ids])&(data<bins[ids+1]))
      new_vec[i,j] = ids
    
    n_bin = len(bins)-1
    new_vec += 2
    return new_vec, bins, n_bin


  def __len__(self):
      return self.n_of_data

  def __getitem__(self, idx):
    return self.generated_sequences_ids[idx], self.generated_sequences_ids_one_hot[idx], self.abs_values[idx]
```

## Model 

```python
class Encoder(nn.Module):
    def __init__(self, input_dim: int, output_vocabulary_size: int, latent_dim: int, hidden_units: int):
        super(Encoder, self).__init__()
        self.lay_1 = nn.Linear(input_dim, hidden_units)
        self.dropout = nn.Dropout(0.2)
        self.sigma_parameter = nn.Linear(hidden_units, latent_dim)
    
    def forward(self, x):
        x = F.relu(self.lay_1(x))
        x = self.dropout(x)
        sigma = F.sigmoid(self.sigma_parameter(x)) 
        return sigma

class Decoder(nn.Module):
    def __init__(self, latent_dim: int, hidden_dim: int, nfft: int, n_vocabulary: int):
        super().__init__()
        self.nfft = nfft
        self.n_vocabulary = n_vocabulary

        self.lay_1 = nn.Linear(latent_dim, hidden_dim)
        self.dropout = nn.Dropout(0.2)
        self.out = nn.Linear(hidden_dim, nfft*n_vocabulary)

    def forward(self, z, sigma):
        z = F.relu(self.lay_1(z))
        z = self.dropout(z)
        out = F.softmax(self.out(z), dim=-1)
        return out.view(-1, self.nfft, self.n_vocabulary)

class VAE(nn.Module):
    def __init__(self, input_sequence_dim: int, latent_dim: int, hidden_dim: int, n_output_vocabulary: int):
        super(VAE, self).__init__()
        self.encoder = Encoder(input_sequence_dim, n_output_vocabulary, latent_dim, hidden_dim)
        self.decoder = Decoder(latent_dim, hidden_dim, input_sequence_dim, n_output_vocabulary)

        self.distribution = torch.distributions.uniform.Uniform(0, 1)
        self.sigma_constant = np.asarray([1]*latent_dim)
    

    def calculate_diffetropy(self, sigma):
       sigma = torch.tensor(sigma).to(device)
       sigma = sigma.detach().cpu().numpy()
      
       return torch.tensor(1 + np.log(sigma/np.sqrt(2)) + np.euler_gamma/2).to(device)
    
    def kl_divergence(self, sigma):
       Entropy_p = self.calculate_diffetropy(self.sigma_constant)
       Entropy_q = self.calculate_diffetropy(sigma)
       kl = Entropy_q - Entropy_p
       return kl.sum(-1)

    def reparametrization(self, sigma):
        sample = self.distribution.sample(sigma.shape).to(device)

        return torch.tensor(np.sqrt(np.log(sample.cpu())*(-2))*sigma.cpu()).to(device)

    def forward(self, x):
        sigma = self.encoder(x)
        z = self.reparametrization(sigma)
        reconstructed_x = self.decoder(z, sigma)
        return reconstructed_x, sigma
```

## Loss function and accuracy function

```python
def loss_function(X, X_one_hot, output_x, x_good_space, sigma, model):
    recon_loss = F.cross_entropy(X_one_hot, output_x)
    supervised_loss = F.cross_entropy(X_one_hot, x_good_space)
    KLD = model.kl_divergence(sigma).mean()
    
    return KLD - recon_loss + recon_loss

def accuracy_function(pred_list, X_list):
    X_list = X_list.permute(1,0,2).to(device)
    accs = torch.tensor([(pred.argmax(1).to(device) == X.argmax(1).to(device)).type(torch.float).sum().item() for pred, X in zip(pred_list, X_list)])
    return torch.mean(accs)
```

## Train and test loop

```python
def train_loop(dataloader, model, loss_critertion, optimizer):
  n_of_batches = len(dataloader)
  size = len(dataloader.dataset)
  loss_container = []

  train_loss = 0.0
  train_accuracy = 0.0

  for id_batch, (X, X_one_hot, y) in enumerate(dataloader):
      X = X.to(device)
      X_one_hot = X_one_hot.to(device)
      y = y.to(device)
      optimizer.zero_grad()
      prediction_reconstructed, sigma = model(X)
      prediction_good_vector = model.decoder(y, sigma)

      loss = loss_critertion(X, X_one_hot, prediction_reconstructed, prediction_good_vector, sigma, model)

      loss.backward()
      optimizer.step()

      train_loss += loss.item()
      train_accuracy += accuracy_function(prediction_good_vector, X_one_hot)

      if id_batch % 1000 == 0:
        loss, current = loss.item(), id_batch * len(X)
        print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
        print()
  
  train_loss /= n_of_batches
  train_accuracy /= size
  
  return train_loss, train_accuracy


def test_loop(dataloader, model, loss_critertion, optimizer):
  size = len(dataloader.dataset)
  num_batches = len(dataloader)
  test_loss, test_accuracy = 0, 0

  with torch.no_grad():
    for X, X_one_hot, abs_val in dataloader:
      pred, sigma = model(X.to(device))
      X_one_hot = X_one_hot.to(device)
      prediction_good_vector = model.decoder(abs_val.to(device), sigma)
      prediction_reconstructed = pred.to(device)
      test_loss += loss_critertion(X, X_one_hot, prediction_reconstructed, prediction_good_vector, sigma, model)
      test_accuracy +=  accuracy_function(prediction_good_vector, X_one_hot)

  test_loss /= num_batches
  test_accuracy /= size
  print(f"Test Error: \n Accuracy: {(100*test_accuracy):>0.1f}%, Avg loss: {test_loss:>8f} \n")
  return test_loss, test_accuracy
```

```python
model = VAE(nfft, latent_dim, hidden_dim, n_output_vocabulary).to(device)

train_losses = []
train_accuracies = []

test_losses = []
test_accuracies = []


optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)


for n_epoch in range(epochs):
    print("-------Training model--------")
    print(f"Epoch {n_epoch+1}\n-------------------------------")
    train_loss, train_accuracy = train_loop(dataloader_train,
                                            model,
                                            loss_function, 
                                            optimizer)

    test_loss, test_accuracy = test_loop(dataloader_train,
                                         model,
                                         loss_function, 
                                         optimizer)


    train_losses.append(train_loss)
    train_accuracies.append(train_accuracy)

    test_losses.append(test_loss)
    test_accuracies.append(test_accuracy)



```
